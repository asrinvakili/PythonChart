# Horizontal BarPlot

نمودار ستونی یا BarPlot یکی از پرکاربردترین و محبوب‌ترین نمودارها در تحلیل داده‌ها است. این نوع نمودار برای نمایش و مقایسه مقادیر مختلفی که به دسته‌های مختلف تعلق دارند استفاده می‌شود. هر ستون ارتفاع یا طولی دارد که نشان دهنده مقدار یا فراوانی آن دسته خاص است.

کاربردهای نمودار ستونی

## مقایسه دسته‌ها

تجزیه و تحلیل داده‌های طبقه‌بندی شده

نمایش تغییرات در طول زمان

بررسی توزیع فراوانی

تجزیه و تحلیل داده‌های دسته‌بندی شده

### نمودار ستونی (BarPlot) پارامترهای مختلفی دارد که با تنظیم آنها می‌توانید نمودار خود را سفارشی کنید. در اینجا پارامترهای اصلی و گزینه‌های ممکن برای هر کدام را با مثال‌های متفاوت توضیح می‌دهم:

### 1. x

گزینه‌ها: دسته‌بندی‌ها یا نام‌های هر ستون که به صورت لیستی از مقادیر داده می‌شوند.

categories = ['A', 'B', 'C', 'D']

### 2. height

گزینه‌ها: مقادیر یا فراوانی‌ها که ارتفاع هر ستون را مشخص می‌کند.

values = [10, 24, 36, 15]

### 3. color

گزینه‌ها: رنگ‌های ستون‌ها. می‌توان از نام رنگ‌ها، کدهای هگزا (hex) یا اعداد RGB استفاده کرد.

color='blue'  # یک رنگ برای همه ستون‌ها
### یا
colors=['blue', 'green', 'red', 'purple']  # رنگ‌های متفاوت برای هر ستون

### 4. edgecolor

گزینه‌ها: رنگ حاشیه ستون‌ها. می‌توان از نام رنگ‌ها، کدهای هگزا (hex) یا اعداد RGB استفاده کرد.

edgecolor='black'

### 5. linewidth

گزینه‌ها: عرض حاشیه ستون‌ها به صورت عددی.

linewidth=1.5

### 6. align

گزینه‌ها: تراز ستون‌ها (وسط یا لبه). گزینه‌های ممکن 'center' و 'edge' هستند.

align='center'  # تراز وسط
### یا
align='edge'  # تراز لبه

### 7. alpha

گزینه‌ها: شفافیت ستون‌ها به صورت عددی بین 0 و 1.

alpha=0.7  # شفافیت 70%

### 8. width

گزینه‌ها: عرض ستون‌ها به صورت عددی.

width=0.8  # عرض ستون‌ها

```python
plt.figure(dpi=80)

sns.barplot(x=[100, 180, 50, 20, 120], y=['C#', 'Python', 'R', 'MatLab', 'JS'], orient='h')

plt.show()
```

### Vertical BarPlot

# Dist Plot

برای نشان دادن فراوانی بازه های مختلف به کار می رود. همینطور برای نشان دادن توزیع احتمالی بسیار کاربرد دارد. این نمودار در ماشین لرنینگ و آمار بسیار معروف است.

نمودار توزیع یا displot یکی از ابزارهای قدرتمند در کتابخانه seaborn برای نمایش توزیع یک داده است. این نمودار ترکیبی از نمودار هیستوگرام و نمودار چگالی هسته‌ای (KDE) است که به شما کمک می‌کند تا به راحتی توزیع داده‌ها و تراکم آنها را بررسی کنید.

### کاربردهای نمودار displot

بررسی توزیع داده‌ها: برای نمایش نحوه توزیع داده‌ها و شناسایی الگوها و ویژگی‌ها.

مقایسه چندین توزیع: می‌توان از آن برای مقایسه توزیع داده‌ها در دسته‌های مختلف استفاده کرد.

شناسایی ناهنجاری‌ها: کمک می‌کند تا ناهنجاری‌ها و نقاط دور افتاده (outliers) را شناسایی کنید.

تحلیل تراکم داده‌ها: استفاده از نمودار چگالی هسته‌ای برای تحلیل تراکم داده‌ها.

### پارامترهای مهم در نمودار displot

data: داده‌های ورودی که به صورت DataFrame یا آرایه (array) می‌تواند باشد.

x: نام ستون داده‌ها برای محور x.

hue: دسته‌بندی داده‌ها بر اساس یک متغیر دیگر.

kind: نوع نمودار (هیستوگرام یا KDE یا هر دو). گزینه‌های موجود 'hist', 'kde' و 'ecdf' هستند.

bins: تعداد سطل‌های هیستوگرام.

height: ارتفاع نمودار.

aspect: نسبت عرض به ارتفاع نمودار.

```python
# یک متغییر نرمال با میانگین و دوییشن مشخص که ازش 1000تا می سازه
data = stats.norm(0, 1).rvs(1000)

plt.figure(dpi=80)

sns.displot(data, kind='kde')

plt.show();
```

# Histogram

نمودار هیستوگرام (Histogram) یکی از پرکاربردترین ابزارها برای نمایش توزیع داده‌ها است. این نمودار نشان می‌دهد که مقادیر یک مجموعه داده در چه بازه‌هایی متمرکز شده‌اند. به بیان ساده، هیستوگرام داده‌ها را به چندین گروه تقسیم می‌کند و سپس تعداد داده‌هایی که در هر گروه قرار دارند را به صورت یک ستون نشان می‌دهد.

کاربرد هیستوگرام:
هیستوگرام برای تحلیل بصری توزیع یک مجموعه داده استفاده می‌شود، به ویژه برای داده‌های عددی. با استفاده از آن می‌توان متوجه شد که داده‌ها در چه بازه‌هایی بیشتر تجمع یافته‌اند، شکل کلی توزیع داده‌ها چیست (مانند نرمال یا چولگی)، و داده‌های پرت احتمالی را شناسایی کرد.

نحوه رسم هیستوگرام در پایتون با matplotlib
برای رسم هیستوگرام، از تابع hist() در کتابخانه matplotlib.pyplot استفاده می‌شود. بیایید به پارامترهای مهم این تابع و نحوه استفاده از آن بپردازیم.

### پارامترهای مهم plt.hist():

### data:

داده‌هایی که می‌خواهیم هیستوگرام آن‌ها را رسم کنیم. باید به صورت لیست یا آرایه ارائه شود.

### bins:

تعیین می‌کند که داده‌ها به چند گروه تقسیم شوند. مقدار پیش‌فرض آن به صورت خودکار توسط matplotlib تنظیم می‌شود، ولی می‌توان آن را به دلخواه تغییر داد. مقدار بیشتر برای bins باعث می‌شود که بازه‌های کوچک‌تر و تعداد ستون‌های بیشتری داشته باشیم، که جزئیات بیشتری را نشان می‌دهد.

مثال:

plt.hist(data, bins=10)  # تقسیم داده‌ها به 10 بازه

### range:

مشخص می‌کند که هیستوگرام در چه بازه‌ای از داده‌ها رسم شود. اگر آن را مشخص نکنیم، به طور پیش‌فرض کل بازه داده‌ها استفاده می‌شود. مثال:

plt.hist(data, bins=5, range=(160, 175))  # بازه بین 160 تا 175 سانتی‌متر

### density:

اگر این پارامتر True باشد، محور عمودی به جای تعداد (frequency) به چگالی (density) تغییر پیدا می‌کند، یعنی مجموع مساحت ستون‌ها برابر ۱ خواهد شد. این حالت معمولاً برای مقایسه داده‌ها با توزیع‌های احتمالی استفاده می‌شود. مثال:

plt.hist(data, bins=5, density=True)

### color:

رنگ ستون‌ها را مشخص می‌کند. مثال:

plt.hist(data, bins=5, color='skyblue')

### edgecolor:

رنگ لبه ستون‌ها را تنظیم می‌کند تا ستون‌ها از هم متمایزتر باشند. مثال:

plt.hist(data, bins=5, color='skyblue', edgecolor='black')

### alpha:

شفافیت ستون‌ها را مشخص می‌کند. مقدار آن بین ۰ تا ۱ است (۰ کاملاً شفاف و ۱ کاملاً مات). مثال:

plt.hist(data, bins=5, alpha=0.7)

### label:

نامی برای داده‌ها تعیین می‌کند که می‌توان از آن برای رسم چندین مجموعه داده در یک هیستوگرام استفاده کرد. مثال:

plt.hist(data, bins=5, color='skyblue', label='Height')

plt.legend()  # نمایش توضیحات

### نکته:
در هیستوگرام، ترتیب و نوع داده‌ها تأثیرگذار است، اما به عنوان مثال برای داده‌های رسته‌ای یا غیر ترتیبی مناسب نیست.

```python
plt.figure(dpi=80)

sns.distplot(data)

plt.show();
```

```python
#رسم چند هیستوگرام در یک نمودار
import matplotlib.pyplot as plt  # کتابخانه matplotlib را برای رسم نمودارها وارد می‌کنیم

# داده‌های گروه اول - شامل قد چند نفر از افراد
data1 = [155, 160, 165, 170, 175]
# داده‌های گروه دوم - شامل قد چند نفر دیگر
data2 = [160, 165, 170, 175, 180]

# رسم هیستوگرام برای گروه اول
plt.hist(data1, bins=5, alpha=0.5, color='blue', label='Group 1')  
# data1 را به عنوان ورودی قرار می‌دهیم
# bins=5 یعنی داده‌ها را به 5 دسته تقسیم می‌کنیم
# alpha=0.5 باعث می‌شود ستون‌ها نیمه‌شفاف باشند (با شفافیت 50%)
# color='blue' رنگ ستون‌ها را آبی می‌کند
# label='Group 1' یک برچسب برای داده‌ها تعیین می‌کند که بعداً در legend نمایش داده می‌شود

# رسم هیستوگرام برای گروه دوم
plt.hist(data2, bins=5, alpha=0.5, color='red', label='Group 2')
# data2 را به عنوان ورودی قرار می‌دهیم
# سایر پارامترها مشابه داده‌های اول است، فقط رنگ ستون‌ها قرمز است و برچسب 'Group 2' برای این داده‌ها در نظر گرفته شده

# تنظیم عنوان محور افقی
plt.xlabel("Height (cm)")  # 'Height (cm)' را به عنوان عنوان محور افقی قرار می‌دهد

# تنظیم عنوان محور عمودی
plt.ylabel("Frequency")  # 'Frequency' را به عنوان عنوان محور عمودی قرار می‌دهد

# نمایش legend برای توضیح داده‌ها
plt.legend()  # نمایش توضیحات برای مشخص کردن داده‌های مربوط به گروه 1 و گروه 2

# نمایش نمودار
plt.show()  # نمایش نمودار به کاربر
```

این کد دو مجموعه داده (قد افراد در دو گروه مختلف) را با استفاده از هیستوگرام روی یک نمودار رسم می‌کند. هر گروه داده‌ها رنگ متفاوتی دارند (آبی و قرمز)، و به دلیل شفافیت 0.5 (با استفاده از alpha)، نمودارها می‌توانند روی هم قرار گیرند و قابل تفکیک باشند

در این نمودار، سه رنگ قابل مشاهده است:

آبی برای گروه اول (Group 1)
قرمز برای گروه دوم (Group 2)
بنفش در ناحیه‌ای که دو گروه با هم همپوشانی دارند.
بخشی که بنفش دیده می‌شود، ترکیبی از رنگ‌های آبی و قرمز است و به دلیل شفافیت (alpha=0.5) ایجاد شده است. این همپوشانی نشان می‌دهد که مقادیر داده‌ها در این محدوده برای هر دو گروه تکرار شده‌اند.

### Stacked Horizontal BarPlot

در این نوع نمودار می توان از یک زمانبندی و دسته بندی به صورت همزمان استفاده کرد

```python
plt.figure(dpi=80)

df = pd.DataFrame({
    'Name': ['Ali', 'Sadegh', 'Mohamad'],
    'Apple': [60, 20, 30],
    'Banana': [20, 40, 20]
})


sns.barplot(x='Name', y='Apple', data=df, label="Apple", color="red", alpha=.5)
sns.barplot(x='Name', y='Banana', bottom=[60, 20, 30], data=df, label="Banana", color="orange", alpha=.5)

plt.legend()
plt.show()
```

این کد مربوط به رسم یک Stacked Bar Chart (نمودار میله‌ای انباشته) است که برای نمایش توزیع مقادیر در یک دسته خاص و مقایسه آنها با دسته‌های دیگر به کار می‌رود. به طور خاص، در این نمودار، مقدار مربوط به هر دسته برای دو نوع میوه (سیب و موز) نمایش داده می‌شود و نشان می‌دهد که هر فرد چقدر سیب و چقدر موز دارد.

### کاربرد نمودار Stacked Bar Chart
نمودار میله‌ای انباشته برای مقایسه مقادیر چندین دسته در یک مجموعه داده استفاده می‌شود. این نمودار به شما امکان می‌دهد تا مقدار کل هر دسته را به همراه ترکیب آن مشاهده کنید. در اینجا، می‌توان مقادیر سیب و موز هر فرد را با هم مقایسه کرد و هم‌زمان ترکیب آنها (مقدار کل میوه) را نیز مشاهده کرد.

### Pie Chart

در نمودار دایره ای میزان هر دسته را به کل نسبت می دهیم. در واقع مساحت زیر کمان میزان فرانی را نشان میدهد. همینطور نمودار دونات یک نوع نموداردایره ای است که وسط آن توخالی است

نمودار دایره‌ای یا Pie Chart یکی از نمودارهای بسیار مفید و محبوب است که برای نمایش سهم نسبی هر جزء از کل مجموعه داده‌ها استفاده می‌شود. این نمودار به خصوص زمانی کاربرد دارد که می‌خواهید نسبت هر بخش به کل را به راحتی به تصویر بکشید.

در زیر توضیحاتی در مورد هر یک از پارامترها و گزینه‌هایی که می‌توانید در نمودار دایره‌ای استفاده کنید آورده‌ام. با هر پارامتر یک مثال کوتاه ارائه شده است:

### 1. labels

گزینه‌ها: لیستی از نام‌ها یا برچسب‌ها که هر بخش را توضیح می‌دهد.

labels = ['سیب', 'موز', 'گیلاس', 'خرمالو']

# Scatter Plot

نمودارهای پراکندگی (Scatterplot) در داده‌نگاری استفاده می‌شوند تا رابطه بین دو متغیر را نشان دهند. در ادامه به برخی از ویژگی‌های مهم نمودارهای پراکندگی و توضیحات مربوط به هرکدام می‌پردازیم:

### X و Y:

X: محور افقی که معمولا نشان‌دهنده متغیر مستقل است.

Y: محور عمودی که معمولا نشان‌دهنده متغیر وابسته است.

### Points (نقاط):

هر نقطه در نمودار پراکندگی نشان‌دهنده یک جفت از مقادیر (x, y) است.

### Size (اندازه):

اندازه نقاط در نمودار می‌تواند نشان‌دهنده متغیر دیگری باشد که به متغیرهای x و y اضافه می‌شود.

### Color (رنگ):

رنگ نقاط می‌تواند برای نمایش دسته‌بندی‌های مختلف یا مقادیر متفاوت استفاده شود.

### Alpha (شفافیت):

میزان شفافیت نقاط که می‌تواند برای جلوگیری از تداخل نقاط با یکدیگر استفاده شود.

### Markers (نشانه‌ها):

نوع علامت‌هایی که برای نشان دادن نقاط استفاده می‌شود (مثلا دایره، مربع، مثلث و غیره).

در کدی که قبلا ارائه دادیم، از ویژگی‌های x, y, size, color و alpha استفاده شد. ویژگی‌های دیگری که می‌توانند اضافه شوند عبارتند از:

### Style (استایل):

مشخص کردن استایل نقاط با استفاده از پارامتر style.

### Hue (رنگ‌آمیزی):

استفاده از پارامتر hue برای دسته‌بندی نقاط بر اساس متغیری دیگر.

### Palette (پالت رنگ):

انتخاب پالت رنگی مخصوص برای نقاط مختلف.

### Legend (راهنما):

افزودن راهنما برای نمایش توضیحات نقاط مختلف.

```python

# تولید داده‌ها
x = np.random.random(100) * 100  # تولید 100 عدد تصادفی بین 0 و 100
y = x ** 2 * np.random.random(100) * 10  # تولید داده‌های y با استفاده از داده‌های x و ضرب آن‌ها در اعداد تصادفی
size = np.arange(1, 101)  # تولید آرایه‌ای از اعداد 1 تا 100 برای اندازه نقاط

# ایجاد نمودار
plt.figure(dpi=80)  # ایجاد یک شکل با وضوح 80 نقطه در اینچ
sns.scatterplot(x=x, y=y, color='red', alpha=0.4, size=size, sizes=(20, 200))  # رسم نمودار پراکندگی با تنظیمات مختلف
plt.xlabel('X Values')  # افزودن برچسب به محور x
plt.ylabel('Y Values')  # افزودن برچسب به محور y
plt.title('Scatter Plot with Varying Point Sizes')  # افزودن عنوان به نمودار
plt.show()  # نمایش نمودار
```

# Line Chart

نمودارهای خطی (Lineplot) در داده‌نگاری برای نمایش تغییرات یک متغیر در طول زمان یا در مقابل متغیر دیگر استفاده می‌شوند. این نمودارها معمولاً برای نمایش روندها یا رفتارهای داده‌ها استفاده می‌شوند. در ادامه به برخی از ویژگی‌های مهم نمودارهای خطی و توضیحات مربوط به هرکدام می‌پردازیم:

### X و Y:

X و Y: فقط مقادیر محورهای افقی و عمودی هستند و معمولاً به صورت آرایه یا لیست از مقادیر استفاده می‌شوند.

### Line (خط):

نوع خط با استفاده از پارامتر linestyle تعیین می‌شود:

'-': خط ممتد (solid line)

'--': خط‌چین (dashed line)

'-.': نقطه-خط‌چین (dash-dot line)

':': خط نقطه‌ای (dotted line)

### Markers (نشانه‌ها):

نوع نشانه‌ها با استفاده از پارامتر marker تعیین می‌شود:

'.': نقطه (point marker)

'o': دایره (circle marker)

'v': مثلث به پایین (triangle down marker)

'^': مثلث به بالا (triangle up marker)

's': مربع (square marker)

'+': بعلاوه (plus marker)

'x': ضربدر (cross marker)

### Color (رنگ):

رنگ خط با استفاده از پارامتر color تعیین می‌شود. می‌توانید از نام رنگ‌ها (مثل 'blue', 'green', 'red') یا کدهای رنگی هگزادسیمال (مثل '#00FF00') استفاده کنید.

### Alpha (شفافیت):

شفافیت خط با استفاده از پارامتر alpha تعیین می‌شود که مقدار بین 0 (کاملاً شفاف) و 1 (کاملاً مات) است.

### Size (اندازه):

ضخامت خط با استفاده از پارامتر linewidth تعیین می‌شود:

مقادیر عددی مثل 1.5, 2, 3 و غیره.

### Hue (رنگ‌آمیزی):

دسته‌بندی نقاط بر اساس متغیر دیگری با استفاده از پارامتر hue:

مثلاً hue='category' که برای دسته‌بندی‌های مختلف رنگ‌های متفاوت استفاده می‌شود.

### Style (سبک):

استفاده از سبک‌های مختلف خطوط برای مجموعه داده‌های مختلف:

پارامتر style می‌تواند استفاده شود تا نوع خط یا نشان‌دهنده‌ها را تعیین کند.

### Legend (راهنما):

راهنما برای نمایش توضیحات خطوط مختلف:

می‌توانید با استفاده از plt.legend() راهنما را به نمودار اضافه کنید و با استفاده از پارامتر title عنوان آن را تنظیم کنید.

```python
x = ['Far', 'Ord', 'Kho', 'Tir', 'Mor', 'Sha']
y = [20, 15, 30, 25, 40, 30]

plt.figure(dpi=80)

sns.lineplot(x, y, color='purple', alpha=0.3)

plt.show()
```

# Heatmap

چه کاربردهایی دارد این نمودار
نمودارهای Heatmap یا نقشه‌های حرارتی کاربردهای فراوانی در تحلیل داده‌ها دارند. این نمودارها به ویژه زمانی مفید هستند که بخواهیم الگوها، همبستگی‌ها و اختلافات را در داده‌های چند‌بعدی شناسایی کنیم.
Heatmap ها به دلیل نمایش بصری و آسان‌فهم الگوها و همبستگی‌ها، ابزاری قدرتمند برای تحلیل داده‌های چند‌بعدی و شناسایی روندها و الگوها در داده‌های پیچیده هستند. 

بسیار خوب، بیایید نگاهی به هرکدام از پارامترهای نمودار Heatmap بیندازیم و ببینیم چه مقادیری می‌توانند بگیرند:

### data:

آرایه‌ای دوبعدی یا DataFrame پانداس که داده‌های ورودی برای ایجاد نمودار هستند.

مثال: np.random.rand(10, 12), dataframe

### vmin و vmax:

مقادیر عددی که محدوده مقیاس رنگ را تعیین می‌کنند.

مثال: vmin=0, vmax=1

### cmap:

رشته‌ای که نشان‌دهنده پالت رنگی مورد استفاده است.

مثال: 'viridis', 'coolwarm', 'plasma', 'inferno'

### annot:

مقدار بولین (True یا False) یا آرایه‌ای که تعیین می‌کند آیا مقادیر داده‌ها در هر خانه نمایش داده شوند یا خیر.

مثال: annot=True, annot=np.random.rand(10, 12)

### fmt:

رشته‌ای که فرمت مقادیر نمایش داده شده را تعیین می‌کند.

مثال: fmt='.2f' (نمایش دو رقم اعشار)، fmt='d' (نمایش به صورت عدد صحیح)

### linewidths:

مقدار عددی که ضخامت خطوط جداکننده خانه‌ها را تعیین می‌کند.

مثال: linewidths=0.5

### linecolor:

رشته‌ای که رنگ خطوط جداکننده خانه‌ها را تعیین می‌کند.

مثال: linecolor='white', linecolor='black'

### cbar:

مقدار بولین (True یا False) که تعیین می‌کند آیا نوار رنگ نمایش داده شود یا خیر.

مثال: cbar=True, cbar=False

### cbar_kws:

دیکشنری که شامل تنظیمات خاص برای نوار رنگ است.

مثال: cbar_kws={'orientation': 'horizontal'} (نوار رنگ افقی)

### xticklabels و yticklabels:

لیستی از رشته‌ها یا مقادیر بولین که برچسب‌های محورهای x و y را تعیین می‌کنند.

مثال: xticklabels=['A', 'B', 'C'], yticklabels=False

### mask:

آرایه‌ای با مقادیر True و False که قسمت‌های True را پنهان می‌کند.

مثال: mask=np.zeros_like(data, dtype=bool), mask=np.tril(np.ones_like(data, dtype=bool))

```python
data = np.array([
    [1, 2, 4, 5, 3],
    [2, 3, 5, 1, 1],
    [2, 1, 3, 2, 1],
    [3, 4, 3, 2, 1]
])

plt.figure(dpi=80)

sns.heatmap(np.corrcoef(data), cmap='mako', annot=True)

plt.show()
```

# BoxPlot

نمودار BoxPlot چه کاربردهایی دارد؟
نمودارهای BoxPlot، که به آنها "box-and-whisker plot" نیز گفته می‌شود، ابزارهای قدرتمندی برای نمایش تصویری توزیع داده‌ها و شناسایی نقاط دورافتاده (outliers) هستند. در ادامه به برخی از کاربردهای مهم نمودارهای BoxPlot اشاره می‌کنم:

1. شناسایی نقاط دورافتاده (Outliers)

2. نمایش خلاصه‌واری داده‌ها

3. مقایسه بین گروه‌ها

4. بررسی تقارن داده‌ها

5. تحلیل تغییرات فصلی

6. تحلیل کیفیت تولید

7. تحلیل آماری

8. پژوهش‌های زیست‌محیطی


نمودار BoxPlot می‌تواند پارامترهای متعددی را بپذیرد که هر یک تنظیمات خاص خود را دارند. در زیر به برخی از پارامترهای مهم و کاربردی و مقادیر ممکن آن‌ها می‌پردازیم و برای هر پارامتر مثالی ارائه می‌دهیم:

### data:

داده‌های ورودی که می‌تواند یک آرایه، DataFrame پانداس، یا لیستی از آرایه‌ها باشد.

مثال: data=np.random.randn(100, 4)

### orient:

جهت نمودار، که می‌تواند 'v' (عمودی) یا 'h' (افقی) باشد.

مثال: orient='v'

### color:

رنگ جعبه‌ها.

مثال: color='blue'

### palette:

پالت رنگی برای نمودار، که می‌تواند یک لیست از رنگ‌ها یا نام یک پالت رنگی باشد.

مثال: palette='Set2'

### width:

عرض جعبه‌ها به صورت نسبی.

مثال: width=0.8

### linewidth:

ضخامت خطوط جعبه‌ها.

مثال: linewidth=2

### whis:

تعیین می‌کند که whiskers (خطوط انتهایی) چقدر از جعبه‌ها فاصله داشته باشند. می‌تواند یک مقدار عددی یا یک دنباله‌ای از دو مقدار باشد.

مثال: whis=1.5

### saturation:

میزان اشباع رنگ‌ها.

مثال: saturation=0.75

### showmeans:

تعیین می‌کند که میانگین داده‌ها نمایش داده شود یا نه.

مثال: showmeans=True

### meanline:

نمایش میانگین به صورت خط.

مثال: meanline=True

### showbox:

تعیین می‌کند که جعبه‌ها نمایش داده شوند یا نه.

مثال: showbox=True

### showfliers:

تعیین می‌کند که نقاط دورافتاده نمایش داده شوند یا نه.

مثال: showfliers=False

### notch:

تعیین می‌کند که جعبه‌ها دارای برجستگی باشند یا نه (برای نمایش آزمون میانگین‌ها).

# Joint Plot

نمودار Jointplot در کتابخانه Seaborn یک ابزار قدرتمند برای نمایش و تحلیل توزیع‌های دو متغیره است. این نمودار ترکیبی از یک نمودار پراکندگی (scatterplot) و نمودارهای توزیع (distribution plots) است که توزیع داده‌ها را در هر محور جداگانه نشان می‌دهد.

### کاربردها و ویژگی‌های Jointplot:
تحلیل همبستگی:

توزیع‌های حاشیه‌ای:


شناسایی نقاط دورافتاده:


گزینه‌های متنوع برای نمایش:

### در اینجا لیستی از پارامترهای مهم jointplot در Seaborn، همراه با توضیحات و مثال‌هایی از مقادیر ممکن آنها آورده شده است:

### x و y:

داده‌های محور افقی و عمودی.

مثال: x=x, y=y

### data:

مجموعه داده‌ها به صورت DataFrame پانداس.

مثال: data=my_dataframe

### kind:

نوع نمودار مرکزی. می‌تواند یکی از مقادیر زیر باشد:

'scatter': نمودار پراکندگی (پیش‌فرض)

'reg': نمودار رگرسیون

'hex': نمودار شش‌ضلعی

'kde': نمودار تخمین تراکم هسته‌ای

'hist': نمودار هیستوگرام

مثال: kind='scatter'

### color:

رنگ نمودار.

مثال: color='blue'

### height:

ارتفاع شکل (نمودار).

مثال: height=8

### ratio:

نسبت اندازه نمودار حاشیه‌ای به نمودار مرکزی.

مثال: ratio=5

### space:

فاصله بین نمودارهای حاشیه‌ای و مرکزی.

مثال: space=0.2

### marginal_kws:

دیکشنری از تنظیمات اضافی برای نمودارهای حاشیه‌ای.

مثال: marginal_kws={'bins': 30, 'kde': True}

### joint_kws:

دیکشنری از تنظیمات اضافی برای نمودار مرکزی.

مثال: joint_kws={'alpha': 0.5}

## انواع نمودارهای Jointplot:
### Scatter (پیش‌فرض):

kind='scatter': نمودار پراکندگی.

### Regression:

kind='reg': نمودار رگرسیون که یک خط رگرسیون به نمودار اضافه می‌کند.

### Hexbin:

kind='hex': نمودار شش‌ضلعی که تراکم نقاط در هر شش‌ضلعی را نشان می‌دهد.

### Kernel Density Estimate (KDE):

kind='kde': نمودار تخمین تراکم هسته‌ای که توزیع پیوسته داده‌ها را نمایش می‌دهد.

### Histogram:

kind='hist': نمودار هیستوگرام.

# Swarm Plot

نمودار Swarmplot در کتابخانه Seaborn یک روش قدرتمند برای نمایش داده‌های دسته‌بندی شده است که نقاط داده‌ها را به گونه‌ای نمایش می‌دهد که از هم فاصله بگیرند و همپوشانی نداشته باشند. این نمودار برای نمایش توزیع داده‌ها در بین دسته‌های مختلف استفاده می‌شود و می‌تواند به شناسایی الگوها، نقاط دورافتاده و تراکم داده‌ها کمک کند.

## کاربردهای Swarmplot:
نمایش توزیع داده‌ها در دسته‌های مختلف

شناسایی نقاط دورافتاده

تحلیل تراکم داده‌ها

مقایسه توزیع‌ها بین دسته‌های مختلف

## پارامترهای مهم Swarmplot:
### x و y:

داده‌های محور افقی و عمودی.

مثال: x='day', y='total_bill'

### data:

مجموعه داده‌ها به صورت DataFrame پانداس.

مثال: data=tips

### hue:

دسته‌بندی داده‌ها برای نمایش رنگ‌های مختلف.

مثال: hue='smoker'

### palette:

پالت رنگی برای نمودار.

مثال: palette='Set2'

### size:

اندازه نقاط.

مثال: size=5

### dodge:

جداسازی دسته‌های مختلف با استفاده از hue.

مثال: dodge=True

### orient:

جهت نمودار، که می‌تواند 'v' (عمودی) یا 'h' (افقی) باشد.

مثال: orient='v'

### linewidth:

ضخامت خطوط نقاط.

مثال: linewidth=1

# رسم نمودار سه‌بعدی با Matplotlib
برای استفاده از Matplotlib برای رسم نمودار سه‌بعدی، به کلاسی به نام Axes3D نیاز داریم که در mpl_toolkits.mplot3d موجود است.

### مثال: رسم نمودار سه‌بعدی خطی و سطحی

### توضیح کد:

np.meshgrid(x, y): یک شبکه دکارتی از نقاط در فضای X و Y ایجاد می‌کند.

Z = np.sin(np.sqrt(X**2 + Y**2)): تابعی را برای محور Z تعریف می‌کند.

ax.plot_surface: برای رسم سطح سه‌بعدی استفاده می‌شود.

cmap='viridis': یک پالت رنگ انتخاب می‌کند.

### پارامترهای دیگر برای رسم سه‌بعدی در Matplotlib:

ax.plot_wireframe(X, Y, Z): برای رسم شبکه‌ای از خطوط بدون سطح پر شده.

ax.scatter(x, y, z): برای رسم نقاط پراکنده سه‌بعدی.

ax.plot(x, y, z): برای رسم خطوط سه‌بعدی.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# ایجاد یک شکل (figure) و فعال کردن Axes3D برای رسم سه‌بعدی
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# ایجاد داده‌ها
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# رسم نمودار سطحی سه‌بعدی
ax.plot_surface(X, Y, Z, cmap='viridis')

# اضافه کردن برچسب‌های محورها
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

Plotly یک کتابخانه‌ی دیگر برای رسم نمودارهای سه‌بعدی است که برای نمودارهای تعاملی بسیار مناسب است.

### توضیح کد:

go.Surface: نمودار سطحی سه‌بعدی را ایجاد می‌کند.

fig.update_layout: ویژگی‌های ظاهری نمودار را تنظیم می‌کند (مثل عنوان، اندازه و حاشیه‌ها).

go.Surface: نمودار سطحی سه‌بعدی را ایجاد می‌کند.

fig.update_layout: ویژگی‌های ظاهری نمودار را تنظیم می‌کند (مثل عنوان، اندازه و حاشیه‌ها).

```python
import plotly.graph_objects as go
import numpy as np

# ایجاد داده‌ها
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# ایجاد نمودار سه‌بعدی سطحی
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title='3D Surface Plot', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
#نمودار پراکنده سه‌بعدی (3D Scatter Plot):
fig = go.Figure(data=[go.Scatter3d(
    x=[1, 2, 3, 4, 5],
    y=[10, 11, 12, 13, 14],
    z=[5, 6, 7, 8, 9],
    mode='markers',
    marker=dict(size=5, color='purple')
)])
fig.show()
```

```python
#نمودار خطی سه‌بعدی (3D Line Plot):
fig = go.Figure(data=[go.Scatter3d(
    x=[1, 2, 3, 4, 5],
    y=[10, 11, 12, 13, 14],
    z=[5, 6, 7, 8, 9],
    mode='lines',
    line=dict(color='green', width=3)
)])
fig.show()
```
